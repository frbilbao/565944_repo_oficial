# Funciones sin parámetros ni retorno

Son funciones que ejecutan una tarea sin esperar ningún argumento y que al finalizar no retornan ningún resultado.

Por ejemplo la función saludar() que muestra en pantalla un mensaje de bienvenida. O la función mostrar_menu() que muestra en pantalla las funcionalidades de nuestra aplicación.

```python
# Declaramos la función saludar
def saludar():
    print("Bienvenido al sistema")
```

```python
# Invocamos a la función saludar para que se ejecute
saludar()
```

```python
# Crear una función que imprima un menú de opciones
def mostrar_menu():
    print("Menú de opciones:")
    print("1. Agregar nombre")
    print("2. Eliminar nombre")
    print("3. Mostrar lista")
    print("4. Salir")
```

```python
# Invocamos a la función mostrar_menu para que se ejecute
mostrar_menu()
```

# Funciones sin parámetros pero con retorno

Son funciones que para realizar una tarea no necesitan argumento, pero que al finalizar si retornan resultados.

Por ejemplo la función leer_nombre(), que caputura y retorna el dato ingresado por consola.

```python
# Creamos una funcion que lea nombres por terminal y los retorne
def leer_nombre():
    nombre = input("Ingrese un nombre: ")
    return nombre
```

```python
# Validamos la función leer_nombre
nombre_leido = leer_nombre()
print("El nombre ingresado es:", nombre_leido)
```

# Funciones con parámetros y sin retorno

Son funciones que para realizar su tarea necesitan datos, pero que al finalizar no retornan ningún resultado. Por ejemplo la función agregar_estudiante(nombre), que recibe el nombre de un estudiante para agregarlo a una lista global.

```python
# Declaramos la lista global de estudiantes
estudiantes = []
```

```python
# Declaramos la función agregar_estudiante que recibe un nombre y lo agrega a la lista de estudiantes
def agregar_estudiante(nombre):
    estudiantes.append(nombre)
    print(f"Estudiante '{nombre}' agregado a la lista.")
```

```python
# Validamos la funcion agregar_estudiante
nuevo_estudiante = leer_nombre()
agregar_estudiante(nuevo_estudiante)
print("Lista de estudiantes:", estudiantes)
```

Comentario: cuando la función espera dos o más parámetros, los mismos pueden pasarse por posición o nombrados. Cuando se elige trabajar por posición, hay que respetar el ordern parámetro-argumento. También se pueden configurar parámetros por defecto. Por últimos es posible usar *args para permitir una cantidad indefinida de argumentos.

# Funciones con parámetros y con retorno

Son funciones que para realizar una tarea esperan un argumento y que al finalizar retornan un resultado. Por ejemplo la función validar_nombre(nombre), que recibe un nombre, lo analiza y retorna True si hay datos o False si esta en blanco.


```python
# Declaramos la función validar_nombre que recibe un nombre y valida que no esté vacío
def validar_nombre(nombre):
    if nombre.strip() == "":
        return False
    return True
```

```python
# Validamos la funcion validar_nombre
nombre = ""
resultado_validacion = validar_nombre(nombre)
print(f"¿El nombre '{nombre}' es válido? {resultado_validacion}")
```

# Bonus track I

Vamos a combinar las funciones creadas para generar un bloque de código con las siguientes características:
1. Imprimir un mensaje de bienvenida
2. Leer el nombre de estudiante por consola
3. Verificar que sea un nombre válido
4. Si es válido guardarlo en la lista global
5. Si es inválido, mostrar un mensaje en la consola

```python
# 1 Imprimimos un mensaje de bienvenida
saludar()
```

```python
# 2. Leer el nombre de estudiante por consola
nuevo_estudiante = leer_nombre()
```

```python
# 3. Verificar que sea un nombre válido
check_nombre = validar_nombre(nuevo_estudiante)
```

```python
# 4. Si es válido guardarlo en la lista global
# 5. Si es inválido, mostrar un mensaje en la consola
if check_nombre:
    agregar_estudiante(nuevo_estudiante)
else:
    print("El nombre ingresado no es válido.")
```

# Bonus track II

Generamos ahora una función que itere la lista global de estudiantes y los muestre numerados en consola

```python
# declaramos al función mostrar_estudiantes
def mostrar_estudiantes(estudiantes):
    print("Lista de estudiantes:")
    for estudiante in estudiantes:
        print(estudiante)
```

```python
# Invocamos a la función mostrar_estudiantes para que se ejecute
mostrar_estudiantes(estudiantes)
```

```python
# mejoramos función mostrar_estudiantes
def mostrar_estudiantes(estudiantes):
    print("Lista de estudiantes:")
    for indice, estudiante in enumerate(estudiantes):
        print(f"{indice + 1}. {estudiante}")
```

```python
# Invocamos a la función mostrar_estudiantes para que se ejecute
mostrar_estudiantes(estudiantes)
```
